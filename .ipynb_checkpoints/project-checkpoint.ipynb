## Splitting the dataset into independent and dependent variables

```python
x, y = newDataFrame.drop("Prediction", axis = 1), newDataFrame["Prediction"]

y = pd.DataFrame(y, columns = ["Prediction"])
x.head(), y.head()
```

## Using sklearn pipeline for Data Preprocessing

```python
numericCols = [col for col in x]

numericTransformer = Pipeline(steps = [
    ("impute", SimpleImputer(strategy = "mean"))
])

preprocessor = ColumnTransformer(transformers = [
    ("numeric", numericTransformer, numericCols)
])

rfc = Pipeline(steps = [
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier())
])

svc = Pipeline(steps = [
    ("preprocessor", preprocessor),
    ("model", SVC())
])

lr = Pipeline(steps = [
    ("preprocessor", preprocessor),
    ("model", LogisticRegression())
])
```

## Dividing the Dataset into Training and Testing sets

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20)
```

## Training Machine Learning Models on the Training Set

```python
rfc.fit(xtrain, ytrain)
```

```python
svc.fit(xtrain, ytrain)
```

```python
lr.fit(xtrain, ytrain)
```

## Creating a Single Function for calculating Performance Metrics for each Model

```python
def calculatePerformanceMetrics(models, ytest, xtest, printResults) -> [[pd.DataFrame]]:

    res0, res1 = [], []

    for model in models:
        
        prec0 = round(precision_score(ytest, models[model].predict(xtest), pos_label = 0), 3)
        rec0 = round(recall_score(ytest, models[model].predict(xtest), pos_label = 0), 3)
        f1_0 = round(f1_score(ytest, models[model].predict(xtest), pos_label = 0), 3)

        prec1 = round(precision_score(ytest, models[model].predict(xtest), pos_label = 1), 3)
        rec1 = round(recall_score(ytest, models[model].predict(xtest), pos_label = 1), 3)
        f1_1 = round(f1_score(ytest, models[model].predict(xtest), pos_label = 1), 3)

        acc = round(accuracy_score(ytest, models[model].predict(xtest)), 3)

        if printResults:
            print(f"\nPerformance Metrics for {model}:")
            print(f"For Class = 0")
            print(f"Precision: {prec0}\nRecall: {rec0}\nF1 Score: {f1_0}\n")
            print(f"For Class = 1")
            print(f"Precision: {prec1}\nRecall: {rec1}\nF1 Score: {f1_1}\n")

            print(f"Mean Accuracy: {acc}")

        res0.append([prec0, rec0, f1_0])
        res1.append([prec1, rec1, f1_1])

    # Create a dataframe and return back the final result
    frame0 = pd.DataFrame(
        res0, index = list(models.keys()), columns = ["Precision", "Recall", "F1 Score"]
    )
    frame1 = pd.DataFrame(
        res1, index = list(models.keys()), columns = ["Precision", "Recall", "F1 Score"]
    )

    return [frame0, frame1]
```

## Analyzing Performance Metrics for each Model

```python
model_dict = {
    "Random Forest Classifier" : rfc,
    "Support Vector Classifier" : svc,
    "Logistic Regression" : lr
}

results = calculatePerformanceMetrics(model_dict, ytest, xtest, printResults = True)
```
